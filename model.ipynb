```python
import os
import cv2
import glob
```

```python
path="C:\\Users\\sadhu\\OneDrive\\Desktop\\research on ml\\dataset\\Fish_Data\\images\\raw_images"
path_f="C:\\Users\\sadhu\\OneDrive\\Desktop\\research on ml\\dataset\\Fish_Data\\images\\classified"
```

```python
images=glob.glob(os.path.join(path,"*"))
array=[]
```

```python
for image in images:
    img=cv2.imread(image)
    name,_=os.path.splitext(os.path.basename(image))
    
    if not(os.path.isdir(os.path.join(path_f,name[:-2]))):
        os.makedirs(os.path.join(os.path.join(path_f,name[:-2])))
        array.append(name[:-2])
    cv2.imwrite(os.path.join(path_f,name[:-2],os.path.basename(image)),img)
    print(os.path.join(path_f,name[:-2],os.path.basename(image)))
```
